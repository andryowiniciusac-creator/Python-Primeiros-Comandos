Teoria 5 - Vídeo #9 - Manipulando textos - 29/03/2026

```python
# Fatiamento de Strings

frase = "Curso Python" # 11 caracteres

print(frase) # Fatia (pega) toda a frase
print(frase[9]) # Fatiamento, fatia a frase
print(frase[6:11])
print(frase[6:12:2]) # Fatia (pega) a cada 2 caracteres
print(frase[:5]) # equivale à [0:5]
print(frase[6:]) # equivale à [6:(último caractere)]
print(frase[0::3])
```

```python
# Análise de Strings

fr = "Fundamentos de Matemática Elementar"
print(len(fr)) # Número total de caracteres
print(fr.count("a")) # Conta quantos "a" tem em fr
print(fr.count("A"))
print(fr.count("a", 6,33)) # Contagem com fatiamento
print(fr.find("os")) # Indica onde está o caractere e onde começou o "os"
print(fr.find("zia")) # Inexistência
print("de" in fr) # Tem "de" em fr?
```

```python
# Transformação de Strings

fr = "Tópicos de Física"
print(fr.replace("Física", "Zero")) # Substitui "Física" em fr por "Zero"
print(fr.upper()) # Tudo em maiúsculo
print(fr.lower()) # Tudo em minúsculo
print(fr.capitalize()) # Só a posição zero fica maiúscula
print(fr.title()) # Na posição n com espaço vazio, (n+1) fica maiúscula
fr1 = "  Guidorizzi e Stewart  "
print(fr1)
print(fr1.strip()) # Tira os espaços vazios "inúteis"
print(fr1.lstrip()) # Remove os da esquerda
print(fr1.rstrip()) # Remove os da direita
```

```python
# Divisão de Strings

fr = "Moysés e Halliday"
print(fr.split())
# Criou uma lista; os espaços vazios são as vírgulas

fr1 = fr.split()
print("-".join(fr1))
# "-" é colocado entre elementos

print("-".join(fr))
# Uma string é uma lista, e "-" é colocado entre elementos
```
