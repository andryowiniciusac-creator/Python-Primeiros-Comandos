```python
p = float(input("Digite o peso em quilograma: "))
h = float(input("Digite a altura em metros: "))
if (p/h**2) < 18.5:
  print("Abaixo do peso")
elif (p/h**2) >= 18.5 and (p/h**2) < 25.0:
  print("Peso ideal")
elif (p/h**2) >= 25 and (p/h**2) < 30:
  print("Sobrepeso")
elif (p/h**2) >= 30 and (p/h**2) < 40:
  print("Obesidade")
elif (p/h**2) >= 40:
  print("Obesidade mórbida")
print("O IMC dessa pessoa é {:.2f}".format(p/h**2))
```
